```python
import librosa
from scipy.io import savemat
```

```python
import os
import glob
```

```python
class ManipulacaoAudio():
    def m4e_to_wav(caminho_entrada_m4a, caminho_saida_mat):
        try:
            print("Lendo o arquivo: {caminho_entrada_m4a}...")
            audio_data, sample_rate = librosa.load(caminho_entrada_m4a, sr=None, mono=False)

            dados_matlab = {
                'sinal_audio': audio_data,
                'fs': sample_rate
            }

            print(f"Salvando dados em: {caminho_saida_mat}...")
            savemat(caminho_saida_mat, dados_matlab)
        
            print("Conversão concluída com sucesso!")
        
        except Exception as e:
            print(f"Ocorreu um erro: {e}")
```
